```python
import numpy as np
import matplotlib.pyplot as plt
x1=[-0.5,0,0.5,1,1.5,2]
x2=[]
x3=[]
for i in range(0,6):
    t1=1.5
    t2=0.5
    w1=1
    w2=1
    y1=(t1/w2)-(x1[i]*(w1/w2))
    x2.append(y1)
    y2=(t2/w2)-(x1[i]*(w1/w2))
    x3.append(y2)
print("W1="+str(w1)+" W2="+str(w2)+" T1="+str(t1)+" T2="+str(t2))
print("x1 x2 Actual Output Desired Output")
d1=[0,0,1,1]
d2=[0,1,0,1]
op=[0,1,1,0]
ap=[]
t=1
for j in range(0,4):
    s=(d1[j]*w1)+(d2[j]*w2)
    if(s==t):
        plt.plot(d1[j],d2[j],"yo")
        ap.append(0)
    else:
        plt.plot(d1[j],d2[j],"gx")
        ap.append(1)
for a in range(0,4):
    print(str(d1[a])+" "+str(d2[a])+" \t "+str(ap[a])+" \t\t"+str(op[a])+"")
plt.plot(x1,x2,"-b")
plt.plot(x1,x3,"-b")
plt.show()
```
